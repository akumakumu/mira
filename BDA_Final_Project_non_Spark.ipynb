#### Importer

---



```python
import numpy as np # Array, Linear Algebra, Fourier Transform, Matric
import pandas as pd # Data Processing, CSV file I/O

from sklearn.metrics.pairwise import cosine_similarity # Algoritma
from sklearn.feature_extraction.text import CountVectorizer # Matrix / Vector Transformer
```

#### Dataset

---



```python
# Tunggu hingga steam.csv muncul di Files
# Mengambil dataset dari github
!wget https://raw.githubusercontent.com/akumakumu/datasets/main/steam.csv
```

```python
# Membaca dataset .csv
df = pd.read_csv('steam.csv', index_col=False)
```

```python
#Cek isi dataset
pd.set_option('display.max_columns', None)
df.head(5)
```

#### Filtering

---



```python
# Cek jumlah baris dan kolom data frame
# (baris, kolom)
df.shape
```

```python
# Mengurangi besar data
# hanya mengambil data yang memiliki positive_rating lebih dari 1000
df = df[df['positive_ratings'] > 1000]
```

```python
df.shape
```

```python
df
```

```python
# Untuk mereset index dan mengurutkannya kembali
# parameter Drop untuk menghapus index yang lama dan diganti dengan index yang baru
df = df.reset_index(drop = True)
df.head(5)
```

#### Feature Select

---



```python
# Mengambil kolom yang akan digunakan untuk rekomendasi
# steamspy_tags = genre menurut komunitas
# name = judul video game
columns = ['steamspy_tags','name']
```

```python
# Cek isi
df[columns].head(5)
```

```python
# Cek apakah ada value yang kosong
df[columns].isnull().values.any()
```

```python
important_features = []
for i in range(0, df.shape[0]): # Looping
    important_features.append(df['steamspy_tags'][i] + ' ' + df['name'][i]) # Menggabungkan isi dari 2 kolom menjadi 1

df['important_features'] = important_features
df['game_no'] = range(0, df.shape[0])
```

```python
# Cek apakah kolom berhasil dibuat
df.head(5)
```

#### Count Vectorizer
---



```python
# Mengubah isi dari kolom important_features kedalam bentuk vector
countVectorizer = CountVectorizer().fit_transform(df['important_features'])
```

```python
 # Cek isi variabel matrik countVectorizer
print(countVectorizer)
```

#### Cosine Similarity

---



```python
# Menghitung Cosine Similarity dari matrik countVectorizer
cosineSimilarity = cosine_similarity(countVectorizer)
```

```python
# Cek isi variabel cosineSimilarity
print(cosineSimilarity)
```

```python
# cek jumlah baris kolom variabel cosineSimilarity
# (baris, kolom)
cosineSimilarity.shape
```

#### Testing

---



##### Masukan judul video game yang disukai

```python
# Judul game di sini (sesuai dengan yang ada di dataset)
title = 'Counter-Strike'
```

##### Magic

```python
# Mencari appid judul game yang disukai
app_id = df[df.name == title]['game_no'].values[0]
print(app_id)
```

```python
# Cek Similarity Score antar judul menurut app_id
# menggunakan Cosine Similarity 
scores = list(enumerate(cosineSimilarity[app_id]))
```

```python
# Cek isi variabel scores
scores
```

```python
# Mengurutkan list skor dengan similarity tertinggi ke terendah
sorted_scores = sorted(scores, key = lambda x:x[1], reverse = True)
sorted_scores = sorted_scores[1:]
sorted_scores
```

```python
# Mengubah tipe data positive_rating menjadi tipedata string
df['positive_ratings'] = df['positive_ratings'].astype(str)
```

#### Evaluasi

---



Menampilkan 10 Rekomendasi Game

```python
i = 0

print('Rekomendasi game untuk', title, 'adalah :\n')

for item in sorted_scores:
    print(df[df.game_no == item[0]]['name'].values[0])
    print('Rating Positif : ' + df[df.game_no == item[0]]['positive_ratings'].values[0])
    print('Kemiripan :',  str(round(item[1]*100)) + '%', '\n')
    i = i+1
    if i>9 :
        break
```
